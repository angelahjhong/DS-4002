```python
import pandas as pd
import numpy as np

df= pd.read_csv('/content/DS-4002/hot-dog-survey-data/Hot-Dog-Form-Fa24.csv' , encoding= ("ISO-8859-1"))
df.head()
```

```python
# Cleaned Data Set
new_df = df.drop(['Name','Start time', 'Completion time', "Email"], axis=1)
new_df.head()
# print(new_df.columns)
# new_df.describe()
```

```python
# Data Dictionary
```

```python
new_df.groupby('Are you a UVA undergraduate student').size()
```

```python
new_df.groupby('What is your major?').size()
```

```python
new_df.groupby('Is a hotdog a sandwich').size()
```

```python
# Exploratory Plots
from matplotlib import pyplot as plt
import seaborn as sns
new_df.groupby('What is your major?').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
data = new_df['Is a hotdog a sandwich'].dropna().value_counts()
plt.figure(figsize=(8, 6))
plt.pie(data, labels=data.index, autopct='%1.1f%%', startangle=140)
plt.title('Is a hotdog a sandwich?')
plt.show()
```
